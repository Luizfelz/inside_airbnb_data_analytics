Linha de raciocínio para confecção do projeto:

- Primeiramente, foram feitas as importações dos CSVs para dentro do databricks, através do "Create Table".

#### Imports

```
from pyspark.sql.functions import when, regexp_replace, col, isnan, isnull

NUM_LINHAS = 3000
```

#### Calendar CSV

```
df_calendar = spark.read.table('default.calendar')
df_calendar = df_calendar.limit(NUM_LINHAS)
df_calendar = df_calendar.dropDuplicates()

df_calendar = df_calendar.withColumn('available', when(df_calendar['available'] == 'f', 0).otherwise(1))
df_calendar = df_calendar.withColumn('price', regexp_replace('price', r'[$,]', ''))
df_calendar = df_calendar.withColumn('adjusted_price', regexp_replace('adjusted_price', r'[$,]', ''))

df_calendar = df_calendar.withColumn('available', col('available').cast('smallint')) # https://docs.databricks.com/en/sql/language-manual/sql-ref-datatypes.html
df_calendar = df_calendar.withColumn('price', col('price').cast('float'))
df_calendar = df_calendar.withColumn('adjusted_price', col('adjusted_price').cast('float'))

df_calendar.limit(5).display()

(
    df_calendar.write.format('jdbc')
        .mode('overwrite')
        .option('driver', 'org.postgresql.Driver')
        .option('url', 'jdbc:postgresql://motty.db.elephantsql.com:5432/ydqpwwmu')
        .option('user', 'ydqpwwmu')
        .option('password', 'N88eKvIl2uW7uULn7IHShwKglRsjb6Qx')
        .option('dbtable', 'tb_calendar_raw')
        .save()
 )
```

#### Reviews CSV

```
df_reviews = spark.read.table("default.reviews")
df_reviews = df_reviews.limit(NUM_LINHAS)
df_reviews = df_reviews.dropDuplicates()

df_reviews = df_reviews.withColumn('comments', regexp_replace('comments', r'<br/>', ''))
df_reviews = df_reviews.withColumn('comments', regexp_replace('comments', r'</b>', ''))
df_reviews = df_reviews.withColumn('comments', regexp_replace('comments', r'<b>', ''))

df_reviews.limit(5).display()

(
    df_reviews.write.format('jdbc')
        .mode('overwrite')
        .option('driver', 'org.postgresql.Driver')
        .option('url', 'jdbc:postgresql://motty.db.elephantsql.com:5432/ydqpwwmu')
        .option('user', 'ydqpwwmu')
        .option('password', 'N88eKvIl2uW7uULn7IHShwKglRsjb6Qx')
        .option('dbtable', 'tb_reviews_raw')
        .save()
 )
```
